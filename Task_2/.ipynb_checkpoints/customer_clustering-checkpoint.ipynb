### Performing EDA

```python
column_name = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
```

```python
def box_plot(customer_data, col_name):
    fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(20,7))
    for i, name in enumerate(col_name):
        sns.boxplot(x=customer_data['gender'], y=customer_data[name], ax=axs[i])
        axs[i].set_title("Gender VS {}".format(name), y=1.0, fontsize=20, color='r')
        axs[i].set_xlabel('Gender', fontsize=12)
        axs[i].set_ylabel('{}'.format(name), fontsize=12)
    fig.tight_layout()
    plt.show()
    
```

```python
box_plot(customer_data, column_name)
```

```python
df = customer_data.copy()
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['gender'], drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scaled_df = scale.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
```

```python
cluster_center = []
cluster = range(1, 12)
for i in cluster:
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_df)
    cluster_center.append(kmeans.inertia_)
```

```python
def elbow_plot(center, cluster):
    fig = plt.subplots(figsize=(15, 8))
    plt.plot(cluster, center, 'go-')
    plt.xlabel("Number of Clusters (K)")
    plt.ylabel("Cluster Inertia")
    plt.title("Elbow Plot of KMeans")
    plt.annotate('', xy=(9,210), xytext=(3,500),arrowprops=dict(arrowstyle= '-|>',
                             color='red',
                             lw=3.5,
                             ls='--'))
    plt.annotate('After Each Iteration Accuracy Increase', xytext=(5,300), xy=(9,200), rotation=-20, fontsize=12, color='blue')
    plt.annotate('', xytext=(5, 0), xy=(5,315), arrowprops=dict(arrowstyle='->', color='blue', lw=3.0, ls='--'))
    plt.annotate('5 is the optimal point', xytext=(5.2,150), xy=(10,100), fontsize=12, color='blue')
    plt.ylim(0)
    plt.xlim(0)
    plt.show()
```

```python
elbow_plot(cluster_center, cluster)
```

# Final Model

```python
new_model = KMeans(n_clusters=5)
new_model.fit(scaled_df)
```

```python
new_model.labels_
```

```python
customer_data['clusters'] = new_model.labels_
```

```python
customer_data.head()
```

# Cluster Diagram

```python
plt.figure(figsize=(10,7))
plt.scatter(x=customer_data['Spending Score (1-100)'], y=customer_data['Annual Income (k$)'], c=customer_data['clusters'])
```
