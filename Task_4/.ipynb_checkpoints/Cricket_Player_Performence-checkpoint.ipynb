### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Batsman_Data.csv')
df.head()
```

```python
df.tail()
```

# Cleaning Data

```python
df.drop(columns=['Unnamed: 0', 'Player_ID', 'Match_ID'], axis=1, inplace=True)
```

### Remove all Special Character from columns

```python
df['Bat1']=df['Bat1'].str.replace('*', '', regex=True)
```

```python
df['SR'] = df['SR'].str.replace('-', '0')
```

```python
df['Opposition'] = df['Opposition'].str.replace('v', '')
```

### Remove All Rows which have DNB and TDNB

```python
index_values = df[(df['Bat1'] == 'DNB') | (df['Bat1'] == 'TDNB') | (df['Bat1'] == 'absent') | (df['Bat1'] == 'sub')].index
df.drop(index_values, axis=0, inplace=True)
df = df.reset_index()
```

```python
df.drop(columns=['index'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# check datatype of each column
df.dtypes
```

```python
# Changing Data types of columns
df['Bat1'] = df['Bat1'].astype('int')
df['Runs'] = df['Runs'].astype('int')
df['BF'] = df['BF'].astype('int')
df['SR'] = df['SR'].astype('float')
df['4s'] = df['4s'].astype('int')
df['6s'] = df['6s'].astype('int')
```

```python
# Change Start Date column to datetime data type
df['Start Date'] = pd.to_datetime(df['Start Date'])
df.head()
```

```python
# Sperate Day, Month and year from Start Date column
df['Day'] = df['Start Date'].map(lambda x: x.day)
df['Month'] = df['Start Date'].map(lambda x: x.month)
df['Year'] = df['Start Date'].map(lambda x: x.year)
df.drop(columns=['Start Date'], axis=1, inplace=True)
```

```python
# Data type of each column
df.dtypes
```

```python
# Null values from data frame
df.isnull().sum()
```

```python
# checking duplicate values
df.duplicated().sum()
```

```python
# Regression plot
col_name = ['BF', 'SR', '4s', '6s']
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(20,10))
axs =axs.flatten()
for i, name in enumerate(col_name):
    sns.regplot(x='Runs', y=name, data=df, ax=axs[i])
    axs[i].set_title("Runs VS {}".format(name), fontsize=20)
    axs[i].set_xlabel("Runs", fontsize=15)
    axs[i].set_ylabel('{}'.format(name), fontsize=15)
fig.tight_layout()
plt.show()
```

```python
col_name = ['Runs','BF', 'SR', '4s', '6s']
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(15,7))
axs =axs.flatten()
for i, name in enumerate(col_name):
    sns.kdeplot(x=name ,data=df, ax=axs[i])
    axs[i].set_title("{} Distribution".format(name), fontsize=20)
    axs[i].set_xlabel("Runs", fontsize=15)
fig.tight_layout()
plt.show()
```

```python
X = df[['Runs', 'BF', 'SR', '4s', '6s']]
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
scaled_x = scale.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
cluster_inertia = []
cluster = range(1,12)
for i in cluster:
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, algorithm='lloyd', random_state=42)
    kmeans.fit(scaled_x)
    cluster_inertia.append(kmeans.inertia_)
```

```python
def elbow_plot(inertia, cluster):
    fig = plt.subplots(figsize=(15, 8))
    plt.plot(cluster, inertia, 'go-')
    plt.xlabel("Number of Clusters (K)")
    plt.ylabel("Cluster Inertia")
    plt.title("Elbow Plot of KMeans")
    plt.ylim(0)
    plt.xlim(0)
    plt.show()
```

```python
elbow_plot(cluster_inertia, cluster)
```

```python
new_model = KMeans(n_clusters=2, init='k-means++', max_iter=300, algorithm='lloyd', random_state=42)
new_model.fit(scaled_x)
```
